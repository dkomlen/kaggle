```python
sample_submission = pd.read_csv('/kaggle/input/tabular-playground-series-feb-2022/sample_submission.csv')
test_raw = pd.read_csv('/kaggle/input/tabular-playground-series-feb-2022/test.csv')
train_raw = pd.read_csv('/kaggle/input/tabular-playground-series-feb-2022/train.csv')
```

```python
X_train = train_raw.drop(labels=['target'], axis=1)
y_train = train_raw['target']
```

```python
model = XGBClassifier()
model.fit(X_train, y_train, verbose=True)
```

```python
y_test = model.predict(test_raw)
```

```python
result = test_raw[['row_id']]
result['target'] = y_test
result
```

```python
result.to_csv('result.csv', index=False)
```
